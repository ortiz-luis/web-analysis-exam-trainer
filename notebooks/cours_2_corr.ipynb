# Exercice 2 : Sélection et filtrage avec `.loc` - CORRECTION

```python
# Données pour l'exercice
entreprise = pd.DataFrame({
    'employe': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'departement': ['Ventes', 'IT', 'Ventes', 'Finance', 'IT', 'Ventes'],
    'salaire': [52000, 65000, 48000, 58000, 72000, 45000],
    'anciennete': [3, 7, 2, 5, 8, 1],
    'evaluation': [4.2, 4.8, 3.5, 4.0, 4.9, 3.8]
}).set_index('employe')

print("DataFrame de base :")
print(entreprise)
```

```python
###################
#   CORRECTION    #
###################

# 1. Affichez le salaire et l'évaluation de Bob
print("=== Salaire et évaluation de Bob ===")
print(entreprise.loc['Bob', ['salaire', 'evaluation']])
```

```python
# 2. Affichez toutes les données des employés du département 'Ventes'
print("=== Employés du département Ventes ===")
print(entreprise.loc[entreprise['departement'] == 'Ventes'])
```

```python
# 3. Affichez les employés avec un salaire > 55000 ET une évaluation >= 4.0
print("=== Salaire > 55k ET évaluation >= 4.0 ===")
condition = (entreprise['salaire'] > 55000) & (entreprise['evaluation'] >= 4.0)
print(entreprise.loc[condition])
```

```python
# 4. Créez une nouvelle colonne 'bonus'
entreprise['bonus'] = 0  # valeur par défaut
entreprise.loc[entreprise['evaluation'] >= 4.5, 'bonus'] = 5000
entreprise.loc[(entreprise['evaluation'] >= 4.0) & (entreprise['evaluation'] < 4.5), 'bonus'] = 2000

print("=== Avec colonne bonus ===")
print(entreprise)
```

# Exercice 3 : Nettoyage et fusion - CORRECTION

```python
# Données pour l'exercice
ventes_brutes = pd.DataFrame({
    'produit_id': ['P001', 'P002', 'P003', 'P001', 'P004'],
    'quantite': [10, None, 15, 8, 20],
    'prix_unitaire': [25.0, 30.0, None, 25.0, 15.0]
})

catalogue = pd.DataFrame({
    'produit_id': ['P001', 'P002', 'P003', 'P004'],
    'nom_produit': ['Widget A', 'Widget B', 'Gadget X', 'Gadget Y'],
    'categorie': ['Widgets', 'Widgets', 'Gadgets', 'Gadgets']
})

print("Ventes brutes :")
print(ventes_brutes)
print("\nCatalogue :")
print(catalogue)
```

```python
###################
#   CORRECTION    #
###################

# 1. Remplacez les NaN
ventes_brutes['quantite'] = ventes_brutes['quantite'].fillna(0)
ventes_brutes['prix_unitaire'] = ventes_brutes['prix_unitaire'].fillna(
    ventes_brutes['prix_unitaire'].mean()
)

print("Après remplissage des NaN :")
print(ventes_brutes)
```

```python
# 2. Fusionnez avec catalogue
ventes_complete = pd.merge(ventes_brutes, catalogue, on='produit_id')

print("Après fusion :")
print(ventes_complete)
```

```python
# 3. Créez 'montant_total'
ventes_complete['montant_total'] = ventes_complete['quantite'] * ventes_complete['prix_unitaire']

print("Résultat final :")
print(ventes_complete)
```

# Exercice 4 : Analyse statistique - CORRECTION

```python
# Générer le dataset
np.random.seed(42)

n = 100
employes_data = pd.DataFrame({
    'employe_id': range(1, n+1),
    'departement': np.random.choice(['Ventes', 'IT', 'Finance', 'Marketing'], n),
    'anciennete': np.random.randint(1, 15, n),
    'salaire': np.random.normal(50000, 10000, n).round(0),
    'performance': np.random.uniform(2.5, 5.0, n).round(1)
})

employes_data.loc[employes_data['departement'] == 'IT', 'salaire'] *= 1.15
employes_data.loc[employes_data['departement'] == 'Finance', 'salaire'] *= 1.10

print("DataFrame :")
print(employes_data.head())
```

```python
###################
#   CORRECTION    #
###################

# 1. Statistiques descriptives
print("=== Statistiques descriptives ===")
print(employes_data.describe())
```

```python
# 2. Salaire moyen par département
print("=== Salaire moyen par département ===")
print(employes_data.groupby('departement')['salaire'].mean().round(0))
```

```python
# 3. Corrélation ancienneté / salaire
corr = employes_data['anciennete'].corr(employes_data['salaire'])
print(f"Corrélation ancienneté-salaire : {corr:.3f}")
```

```python
# 4. Nombre d'employés par département
print("=== Effectif par département ===")
print(employes_data.groupby('departement')['employe_id'].count())
```

```python
# 5. Performance moyenne si salaire > 55000
perf_moyenne = employes_data.loc[employes_data['salaire'] > 55000, 'performance'].mean()
print(f"Performance moyenne (salaire > 55k) : {perf_moyenne:.2f}")
```

# Exercice Final : Cas pratique complet - CORRECTION

```python
# Génération des données
np.random.seed(123)

n_produits = 50
categories = ['Électronique', 'Vêtements', 'Maison', 'Sport', 'Beauté']

produits_ecommerce = pd.DataFrame({
    'produit_id': [f'P{str(i).zfill(3)}' for i in range(1, n_produits+1)],
    'categorie': np.random.choice(categories, n_produits),
    'prix': np.random.uniform(10, 500, n_produits).round(2),
    'cout_achat': np.random.uniform(5, 250, n_produits).round(2),
    'stock': np.random.randint(0, 200, n_produits),
    'ventes_mois': np.random.randint(0, 100, n_produits),
    'note_client': np.random.uniform(2.5, 5.0, n_produits).round(1)
})

# Ajouter quelques NaN
produits_ecommerce.loc[5, 'note_client'] = None
produits_ecommerce.loc[15, 'cout_achat'] = None
produits_ecommerce.loc[25, 'ventes_mois'] = None

df = produits_ecommerce.copy()  # Travail sur une copie
print("DataFrame original :")
print(df.head(10))
```

```python
###################
#   CORRECTION    #
###################

# 1. EXPLORATION
# a) Dimensions et types
print("=== 1a. Dimensions ===")
print(f"Shape : {df.shape}")
print(f"\nTypes :\n{df.dtypes}")
```

```python
# b) Valeurs manquantes
print("=== 1b. Valeurs manquantes ===")
print(df.isna().sum())
```

```python
# 2. NETTOYAGE
# a) note_client : médiane
df['note_client'] = df['note_client'].fillna(df['note_client'].median())

# b) cout_achat : moyenne
df['cout_achat'] = df['cout_achat'].fillna(df['cout_achat'].mean())

# c) Supprimer lignes où ventes_mois est NaN
df = df.dropna(subset=['ventes_mois'])

print("=== 2. Après nettoyage ===")
print(f"Shape : {df.shape}")
print(f"NaN restants : {df.isna().sum().sum()}")
```

```python
# 3. CALCULS
# a) Marge
df['marge'] = df['prix'] - df['cout_achat']

# b) Taux de marge
df['taux_marge'] = (df['marge'] / df['prix'] * 100).round(1)

# c) CA
df['chiffre_affaires'] = df['prix'] * df['ventes_mois']

print("=== 3. Avec colonnes calculées ===")
print(df[['produit_id', 'prix', 'cout_achat', 'marge', 'taux_marge', 'chiffre_affaires']].head())
```

```python
# 4a. Produits avec taux de marge > 50%
print("=== 4a. Taux de marge > 50% ===")
print(df.loc[df['taux_marge'] > 50, ['produit_id', 'categorie', 'taux_marge']])
```

```python
# 4b. Électronique avec note >= 4.0
print("=== 4b. Électronique avec note >= 4.0 ===")
condition = (df['categorie'] == 'Électronique') & (df['note_client'] >= 4.0)
print(df.loc[condition, ['produit_id', 'prix', 'note_client']])
```

```python
# 4c. Statut stock
df['statut_stock'] = 'OK'
df.loc[df['stock'] == 0, 'statut_stock'] = 'Rupture'
df.loc[(df['stock'] > 0) & (df['stock'] < 20), 'statut_stock'] = 'Faible'

print("=== 4c. Répartition des statuts de stock ===")
print(df['statut_stock'].value_counts())
```

```python
# 5a. CA total et moyen par catégorie
print("=== 5a. CA par catégorie ===")
ca_par_cat = df.groupby('categorie')['chiffre_affaires'].agg(['sum', 'mean']).round(2)
ca_par_cat.columns = ['CA_total', 'CA_moyen']
print(ca_par_cat)
```

```python
# 5b. Catégorie avec meilleur taux de marge moyen
marge_par_cat = df.groupby('categorie')['taux_marge'].mean().round(1)
meilleure_cat = marge_par_cat.idxmax()
print(f"=== 5b. Meilleur taux de marge : {meilleure_cat} ({marge_par_cat[meilleure_cat]}%) ===")
print(marge_par_cat.sort_values(ascending=False))
```

```python
# 5c. Corrélation note / ventes
corr_note_ventes = df['note_client'].corr(df['ventes_mois'])
print(f"=== 5c. Corrélation note_client / ventes_mois : {corr_note_ventes:.3f} ===")
```

# Récapitulatif

| Opération | Méthode |
|-----------|--------|
| Créer un DataFrame | `pd.DataFrame(dict)` ou `pd.DataFrame(liste, columns=[...])` |
| Importer CSV | `pd.read_csv('fichier.csv')` |
| Exporter CSV | `df.to_csv('fichier.csv', index=False)` |
| **Sélectionner (TOUJOURS utiliser)** | **`df.loc[lignes, colonnes]`** |
| Filtrer | `df.loc[df['col'] > valeur]` |
| Modifier | `df.loc[condition, 'col'] = nouvelle_valeur` |
| Gérer les NaN | `df.dropna()` ou `df.fillna(valeur)` |
| Renommer | `df.rename(columns={'ancien': 'nouveau'})` |
| Fusionner | `pd.merge(df1, df2, on='col')` |
| Statistiques | `df.describe()`, `df.mean()`, `df.groupby()` |
