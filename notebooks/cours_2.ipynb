# Partie 4 : Sélection de données avec `.loc` (IMPORTANT)

**`.loc`** est LA méthode recommandée pour sélectionner et modifier des données.

**Syntaxe** : `df.loc[sélection_lignes, sélection_colonnes]`

**Avantages de `.loc` :**
- Fonctionne par **labels** (noms des index/colonnes)
- Permet de **modifier** les données sans warning
- Syntaxe claire et explicite

## Exemple 4.1 : Sélection de colonnes avec `.loc`

```python
# Rappel de notre DataFrame
print(df_ventes_idx)
print()
```

```python
# Sélectionner UNE colonne (: = toutes les lignes)
print("=== Une colonne ===")
print(df_ventes_idx.loc[:, 'ventes_q1'])
print()

# Sélectionner PLUSIEURS colonnes
print("=== Plusieurs colonnes ===")
print(df_ventes_idx.loc[:, ['ventes_q1', 'ventes_q2']])
```

## Exemple 4.2 : Sélection de lignes avec `.loc`

```python
# Sélectionner UNE ligne par son index
print("=== Ligne 'Nord' ===")
print(df_ventes_idx.loc['Nord'])
print()

# Sélectionner PLUSIEURS lignes
print("=== Lignes 'Nord' et 'Sud' ===")
print(df_ventes_idx.loc[['Nord', 'Sud']])
```

## Exemple 4.3 : Sélection combinée lignes + colonnes

```python
# Sélectionner une cellule précise
print("Ventes Q1 de la région Nord :", df_ventes_idx.loc['Nord', 'ventes_q1'])

# Sélectionner un sous-ensemble
print("\n=== Nord et Sud, colonnes ventes ===")
print(df_ventes_idx.loc[['Nord', 'Sud'], ['ventes_q1', 'ventes_q2']])
```

## Exemple 4.4 : Filtrage avec conditions (le plus utile !)

```python
# Régions avec ventes Q1 > 120000
print("=== Ventes Q1 > 120000 ===")
print(df_ventes_idx.loc[df_ventes_idx['ventes_q1'] > 120000])
print()

# Filtrer ET sélectionner certaines colonnes
print("=== Ventes Q1 > 120000, colonnes ventes uniquement ===")
print(df_ventes_idx.loc[df_ventes_idx['ventes_q1'] > 120000, ['ventes_q1', 'ventes_q2']])
```

```python
# Conditions multiples avec & (ET) et | (OU)
# ATTENTION : utiliser des parenthèses autour de chaque condition

# Ventes Q1 > 100000 ET nb_employes > 35
condition = (df_ventes_idx['ventes_q1'] > 100000) & (df_ventes_idx['nb_employes'] > 35)
print("=== Ventes Q1 > 100k ET employés > 35 ===")
print(df_ventes_idx.loc[condition])
print()

# Ventes Q1 > 170000 OU nb_employes < 30
condition2 = (df_ventes_idx['ventes_q1'] > 170000) | (df_ventes_idx['nb_employes'] < 30)
print("=== Ventes Q1 > 170k OU employés < 30 ===")
print(df_ventes_idx.loc[condition2])
```

## Exemple 4.5 : Modifier des données avec `.loc`

**C'est ici que `.loc` est indispensable !**

```python
# Créons une copie pour ne pas modifier l'original
df_modif = df_ventes_idx.copy()

# Modifier une cellule
df_modif.loc['Nord', 'ventes_q1'] = 160000
print("Après modification d'une cellule :")
print(df_modif.loc['Nord'])
```

```python
# Modifier plusieurs valeurs selon une condition
# Exemple : augmenter de 10% les ventes Q2 des régions où Q1 < 120000

df_modif = df_ventes_idx.copy()
print("Avant modification :")
print(df_modif[['ventes_q1', 'ventes_q2']])

# La modification avec .loc
df_modif.loc[df_modif['ventes_q1'] < 120000, 'ventes_q2'] *= 1.10

print("\nAprès +10% sur Q2 si Q1 < 120000 :")
print(df_modif[['ventes_q1', 'ventes_q2']])
```

## Exemple 4.6 : Créer de nouvelles colonnes avec `.loc`

```python
df_calc = df_ventes_idx.copy()

# Créer une colonne calculée (total ventes)
df_calc['ventes_total'] = df_calc['ventes_q1'] + df_calc['ventes_q2']

# Créer une colonne conditionnelle avec .loc
df_calc['performance'] = 'Standard'  # valeur par défaut
df_calc.loc[df_calc['ventes_total'] > 300000, 'performance'] = 'Excellent'
df_calc.loc[df_calc['ventes_total'] < 220000, 'performance'] = 'À améliorer'

print(df_calc)
```

# Partie 5 : Nettoyage des données

Les données réelles contiennent souvent des valeurs manquantes, des doublons, ou des erreurs.

## Exemple 5.1 : Gestion des valeurs manquantes (NaN)

```python
# DataFrame avec des valeurs manquantes
df_manquant = pd.DataFrame({
    'produit': ['A', 'B', 'C', 'D', 'E'],
    'prix': [100, None, 150, 200, None],
    'stock': [50, 30, None, 45, 60],
    'categorie': ['X', 'Y', 'X', None, 'Y']
})

print("DataFrame avec valeurs manquantes :")
print(df_manquant)
print(f"\nNombre de NaN par colonne :")
print(df_manquant.isna().sum())
```

```python
# Option 1 : Supprimer les lignes avec NaN
df_sans_nan = df_manquant.dropna()
print("Après dropna() :")
print(df_sans_nan)
```

```python
# Option 2 : Remplir les NaN avec une valeur
df_rempli = df_manquant.copy()
df_rempli['prix'] = df_rempli['prix'].fillna(df_rempli['prix'].mean())  # moyenne
df_rempli['stock'] = df_rempli['stock'].fillna(0)  # valeur fixe
df_rempli['categorie'] = df_rempli['categorie'].fillna('Inconnu')  # texte

print("Après remplissage :")
print(df_rempli)
```

# Partie 6 : Fusion de DataFrames (merge)

Comme les jointures SQL, `merge` permet de combiner des tables.

## Exemple 6.1 : Merge simple

```python
# Table des commandes
commandes = pd.DataFrame({
    'commande_id': [1, 2, 3, 4, 5],
    'client_id': [101, 102, 101, 103, 102],
    'montant': [250, 180, 320, 95, 410]
})

# Table des clients
clients = pd.DataFrame({
    'client_id': [101, 102, 103, 104],
    'nom': ['Dupont', 'Martin', 'Bernard', 'Petit'],
    'ville': ['Paris', 'Lyon', 'Marseille', 'Bordeaux']
})

print("Commandes :")
print(commandes)
print("\nClients :")
print(clients)
```

```python
# Fusionner sur client_id
df_merged = pd.merge(commandes, clients, on='client_id')
print("Après merge :")
print(df_merged)
```

## Exemple 6.2 : Types de merge

```python
# inner (par défaut) : garde seulement les correspondances
# left : garde toutes les lignes de gauche
# right : garde toutes les lignes de droite
# outer : garde tout

# Left merge : garde tous les clients même sans commande
df_left = pd.merge(clients, commandes, on='client_id', how='left')
print("Left merge (tous les clients) :")
print(df_left)
```

# Partie 7 : Statistiques descriptives

## Exemple 7.1 : Statistiques de base

```python
# describe() pour un aperçu complet
print("=== Statistiques descriptives ===")
print(df_ventes.describe())
```

```python
# Statistiques individuelles
print(f"Moyenne des ventes Q1 : {df_ventes['ventes_q1'].mean():,.0f}€")
print(f"Médiane des ventes Q1 : {df_ventes['ventes_q1'].median():,.0f}€")
print(f"Écart-type des ventes Q1 : {df_ventes['ventes_q1'].std():,.0f}€")
print(f"Min/Max : {df_ventes['ventes_q1'].min():,} - {df_ventes['ventes_q1'].max():,}€")
print(f"Somme totale : {df_ventes['ventes_q1'].sum():,}€")
```

## Exemple 7.2 : Agrégation avec groupby

```python
# Données de transactions
transactions = pd.DataFrame({
    'date': ['2024-01-15', '2024-01-15', '2024-01-16', '2024-01-16', '2024-01-17'],
    'categorie': ['Électronique', 'Vêtements', 'Électronique', 'Alimentation', 'Vêtements'],
    'montant': [599, 89, 149, 45, 120],
    'quantite': [1, 2, 3, 5, 1]
})

# Grouper par catégorie et calculer des statistiques
par_categorie = transactions.groupby('categorie').agg({
    'montant': ['sum', 'mean', 'count'],
    'quantite': 'sum'
})

print("Statistiques par catégorie :")
print(par_categorie)
```

```python
# Groupby simple avec une seule fonction
print("\nMontant total par catégorie :")
print(transactions.groupby('categorie')['montant'].sum())
```

## Exemple 7.3 : Corrélations

```python
# Matrice de corrélation
print("=== Matrice de corrélation ===")
colonnes_num = ['ventes_q1', 'ventes_q2', 'nb_employes', 'budget_marketing']
print(df_ventes[colonnes_num].corr().round(2))
```

```python
# Corrélation entre deux colonnes spécifiques
corr = df_ventes['ventes_q1'].corr(df_ventes['nb_employes'])
print(f"Corrélation ventes Q1 / nb employés : {corr:.3f}")
```

# Récapitulatif

| Opération | Méthode |
|-----------|--------|
| Créer un DataFrame | `pd.DataFrame(dict)` ou `pd.DataFrame(liste, columns=[...])` |
| Importer CSV | `pd.read_csv('fichier.csv')` |
| Exporter CSV | `df.to_csv('fichier.csv', index=False)` |
| **Sélectionner (TOUJOURS utiliser)** | **`df.loc[lignes, colonnes]`** |
| Filtrer | `df.loc[df['col'] > valeur]` |
| Modifier | `df.loc[condition, 'col'] = nouvelle_valeur` |
| Gérer les NaN | `df.dropna()` ou `df.fillna(valeur)` |
| Renommer | `df.rename(columns={'ancien': 'nouveau'})` |
| Fusionner | `pd.merge(df1, df2, on='col')` |
| Statistiques | `df.describe()`, `df.mean()`, `df.groupby()` |
